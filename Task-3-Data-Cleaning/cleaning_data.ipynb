```python
# Pandas aur NumPy import kar rahe hain
# NumPy numerical operations ke liye use hota hai
import pandas as pd
import numpy as np

# Dataset dobara load kar rahe hain
df = pd.read_excel(
    r"C:\Users\vaiva\OneDrive\Desktop\RD-INFRO-DATA-SCIENCE\Dataset\Telco_customer_churn.xlsx"
)
```

```python
# Step 1: Missing values check kar rahe hain
# Kaunse column mein kitne null values hain
df.isnull().sum()
```

```python
# Step 2: Total Charges column ko numeric bana rahe hain
# Kuch values string format mein hoti hain, isliye convert karna zaroori hai
df["Total Charges"] = pd.to_numeric(df["Total Charges"], errors="coerce")
```

```python
# Step 3: Numeric columns identify kar rahe hain
# Sirf numeric columns mein median fill karenge
numeric_columns = df.select_dtypes(include=np.number).columns

# Step 4: Missing values ko median se fill kar rahe hain
# Median outliers se kam affect hota hai
df[numeric_columns] = df[numeric_columns].fillna(
    df[numeric_columns].median()
)
```

```python
# Step 5: Irrelevant columns remove kar rahe hain
# CustomerID aur Lat Long prediction ke kaam ke nahi hain
df.drop(columns=["CustomerID", "Lat Long"], inplace=True)
```

```python
# Step 6: Categorical columns ko clean kar rahe hain
# Extra spaces aur uppercase-lowercase issues remove kar rahe hain
categorical_columns = df.select_dtypes(include="object").columns
for col in categorical_columns:
    df[col] = df[col].str.strip().str.lower()
```
