```python
# SQLite aur Pandas import kar rahe hain
import sqlite3
import pandas as pd
```

```python
# Database se connection establish kar rahe hain
# Ye bilkul wahi hai jo PDF mein diya gaya hai
conn = sqlite3.connect("db.sqlite")
```

```python
# Demo table create kar rahe hain
conn.execute("""
CREATE TABLE IF NOT EXISTS users (
    user_id INTEGER,
    age INTEGER,
    churn INTEGER
)
""")
```

```python
# Sample data insert kar rahe hain
conn.execute("INSERT INTO users VALUES (1, 30, 0)")
conn.execute("INSERT INTO users VALUES (2, 45, 1)")
conn.commit()
```

```python
# SQL query ke through data read kar rahe hain
# EXACTLY like PDF example
df_sql = pd.read_sql("SELECT * FROM users", conn)
```
